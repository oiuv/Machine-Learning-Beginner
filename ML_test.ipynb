```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子以保证可重复性
np.random.seed(0)

# 生成100个样本，每个样本有两个特征，取值范围在0到10之间
X = 10 * np.random.rand(100, 2)

# 根据一个线性模型生成标签，并加上随机噪声
# y = 1 if 3*x1 + 2*x2 > 25, else y = 0
y = (3 * X[:, 0] + 2 * X[:, 1] > 25).astype(int)

# 可视化生成的数据集
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1')
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.show()
```

```python
# 创建一个新的图形对象和3D坐标轴
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 可视化生成的数据集
ax.scatter(X[y == 0][:, 0], X[y == 0][:, 1], y[y == 0], color='blue', label='Class 0')
ax.scatter(X[y == 1][:, 0], X[y == 1][:, 1], y[y == 1], color='red', label='Class 1')

# 设置轴标签
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$y$")
ax.legend()

# 显示图形
plt.show()
```
